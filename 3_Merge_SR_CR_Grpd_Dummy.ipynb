### Merge Groupby and Dummy Service Request and Crime datasets

```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
#import 2019 Service Request dataset which has been grouped by month_created and PolicePrecinct and dummied
dfsr_d = pd.read_csv(r'C:\Documents\projects\HackLA\311\data\WorkedonData\sr_2019_grpby_dummies2.csv', index_col=0)
```
