```python
# Import the necessary packages
import pandas as pd
import os
from datetime import datetime
```

```python
os.getcwd() # gives you the path to the current working directory

# This will change the directory 
os.chdir('C:/Users/AdithiPriya/Desktop/Hack for LA/Geospatial analysis')

# Since the 311 request data file is so huge, we set the low_memory = False...
la_311 = pd.read_csv('MyLA311_Service_Request_Data_2020.csv', low_memory=False) 
crime_data = pd.read_csv('Crime_Data_from_2020_to_Present.csv')
```

```python
# Filter out the bulky items request type.
la_311_filtered = la_311[la_311['RequestType']== 'Bulky Items']
la_311_filtered.head(2)
```

```python
# Since both the dataframes have many column, let us remove the unnecessary ones...
la_311_filtered=la_311_filtered[['SRNumber','CreatedDate', 'UpdatedDate', 'NC', 'NCName', 'RequestType','RequestSource','Address','ZipCode', 'Latitude', 'Longitude', 'Location', 'PolicePrecinct' ]]
la_311_filtered.head(10)
```

```python
# There are 2 date columns in the crime_data df (dataframe). So I am going to choose 
# DATE OCC column and filter the year 2020 using that column...

# Convert the 'DATE OCC' string type to datetime and then filter the year 2020

crime_data['year']= pd.to_datetime(crime_data['DATE OCC']).dt.year 

crime_data_2020 =crime_data[crime_data['year']==2020]
crime_data_2020.head(10)
```

```python
# Renaming the column. Note: you can also use this- pd.merge(df1, df2, left_on='col_name', right_on='col_name') without having 
# to rename but you will end up with 2 columns. This is the cleaner way of doing it.
crime_data_2020 = crime_data_2020.rename(columns= {'AREA NAME': 'PolicePrecinct'})

crime_data_2020.head(10) 
```

```python
# Removinge the unnecessary columns from crime_data_2020...You can redo this depending upon how you want to go with the analysis...

crime_data_2020= crime_data_2020[['DR_NO', 'Date Rptd', 'DATE OCC', 'TIME OCC', 'AREA','Crm Cd Desc', 'Premis Desc', 'PolicePrecinct','LOCATION', 'LAT', 'LON', 'year' ]]
crime_data_2020.head()
```
